# Setting up

```python
# activiate inline plotting
%matplotlib inline
from setup import *
```

```python
#import data
#the data used in this study have been pre-converted into an hdf5 file. 
#to preprocess your own data, please use limix command line brinary
file_name = tutorial_data.get_file('arab107')
geno_reader  = gr.genotype_reader_tables(file_name)
pheno_reader = phr.pheno_reader_tables(file_name)

#the data object allows to query specific genotype or phenotype data
dataset = data.QTLData(geno_reader=geno_reader,pheno_reader=pheno_reader)
```

```python
#flowering phenotypes
phenotype_names = ['5_FT10']
phenotype_ID =  '5_FT10'
```

```python
#load first flowering phenotype
phenotype_query = "(phenotype_ID in %s)" %  str(phenotype_names)

data_subsample = dataset.subsample_phenotypes(phenotype_query=phenotype_query,
                                           intersection=True)

sample_relatedness_unnormalized = data_subsample.getCovariance(normalize=True,center=True)

sample_relatedness  = sample_relatedness_unnormalized/ \
    sample_relatedness_unnormalized.diagonal().mean()

if 1:
    #subsample for speed?
    Is = sp.arange(dataset.geno_pos.shape[0])
    #take every 10th SNP
    Is = Is[::10]
    data_subsample = data_subsample.subsample(cols_geno = Is)

    
#get variables we need from data
snps = data_subsample.getGenotypes(center=False,unit=False)
snps-= snps.mean(axis=0)
snps/= snps.std(axis=0) 

position = data_subsample.getPos()
position,chromBounds = data_util.estCumPos(position=position,offset=100000)

phenotypes,sample_idx = data_subsample.getPhenotypes(phenotype_query=phenotype_query,intersection=True) 
phenotype_std = preprocess.rankStandardizeNormal(phenotypes.values)

N = snps.shape[0]
S = snps.shape[1]
P = phenotypes.shape[1]
print "loaded %d samples, %d phenotypes, %s snps" % (N,P,S)
```

# Linear mixed models

```python
#run a linear mixed model and a standard linear model for comparison
lmm = qtl.test_lmm(snps=snps,pheno=phenotype_std, K=sample_relatedness)
pvalues_lmm = lmm.getPv()[0] 
```

# Lasso 

```python
# running cross-validation
alphas = 2.**(sp.linspace(-7,-2,10))
alphas = alphas[::-1]

# normalize SNPs
lasso = Lasso(warm_start=True,fit_intercept=False)
MSE_train,MSE_test,W_nonzero = lmmlasso.runCrossValidation(lasso,snps,phenotype_std,alphas,n_folds=10,verbose=True)
```

```python
# plotting cross-validation results
idx = sp.argmin(MSE_test.mean(axis=0))

pl.figure(figsize=[20,4])
plt = pl.subplot(1,3,1)
plt.plot(sp.log2(alphas),MSE_train.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('training error')
pl.grid(True)

plt = pl.subplot(1,3,2)
plt.plot(sp.log2(alphas),MSE_test.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('test error')
pl.grid(True)

plt = pl.subplot(1,3,3)
plt.plot(sp.log2(alphas),W_nonzero.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('number of nonzero coefficients')
pl.grid(True)
```

#  LMM-Lasso

```python
# running cross-validation
alphas = 2.**(sp.linspace(-7,-2,10))
alphas = alphas[::-1]

# running LMM-Lasso
lasso = lmmlasso.LmmLasso(warm_start=True,fit_intercept=False)
MSE_train,MSE_test,W_nonzero = lmmlasso.runCrossValidation(lasso,snps,phenotype_std,alphas,n_folds=10,K=sample_relatedness,verbose=True)
```

```python
# plotting cross-validation results
idx = sp.argmin(MSE_test.mean(axis=0))

pl.figure(figsize=[20,4])
plt = pl.subplot(1,3,1)
plt.plot(sp.log2(alphas),MSE_train.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('training error')
pl.grid(True)

plt = pl.subplot(1,3,2)
plt.plot(sp.log2(alphas),MSE_test.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('test error')
pl.grid(True)

plt = pl.subplot(1,3,3)
plt.plot(sp.log2(alphas),W_nonzero.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('number of nonzero coefficients')
pl.grid(True)
```

```python
# plotting weight vector
alpha_cv = alphas[idx]
lasso = lmmlasso.LmmLasso(warm_start=False,fit_intercept=False)
lasso.set_params(alpha=alpha_cv)
lasso = lasso.fit(snps,phenotype_std,K=sample_relatedness)
weights = lasso.coef_
pl.figure(figsize=[20,4])
plt = pl.subplot(2,1,1)
pl.title('LMM-Lasso Weights')
plt.plot(position['pos_cum'],weights,color='k')
pl.xticks([])
pl.ylabel('weights')
plt = pl.subplot(2,1,2)
pl.title('Manhattan Plot')
plot_manhattan(position['pos_cum'],pvalues_lmm,chromBounds)
```

# Genome-Wide Association Studies

```python
# train lmm-lasso on all but one chromsome, estimate p-values on the remaining one
idx = sp.argmin(MSE_test.mean(axis=0))
alpha_cv = alphas[idx]
lasso = lmmlasso.LmmLasso(warm_start=False,fit_intercept=False)
lasso.set_params(alpha=alpha_cv)

chromU = sp.unique(position.chrom)
pv_lmmlasso = sp.zeros(S)
for ichrom in chromU:
    ifg = sp.array(position.chrom==ichrom)
    Xfg = snps[:,ifg]
    Xbg = snps[:,~ifg]
    pv_lmmlasso[ifg] = lmmlasso.testLassoBackgroundModel(lasso,Xfg,Xbg,phenotype_std,use_Kgeno=True)
    
pl.figure(figsize=[20,4])
pl.subplots_adjust(hspace=0.5)
plt = pl.subplot(2,2,1)
plot_manhattan(position['pos_cum'],pv_lmmlasso,chromBounds)
pl.title('LMM (using the LMM-Lasso to estimate the background model)')
plt = pl.subplot(2,2,2)
pl.title('LMM (using the LMM-Lasso to estimate the background model)')
qqplot(pv_lmmlasso)

plt = pl.subplot(2,2,3)
plot_manhattan(position['pos_cum'],pvalues_lmm,chromBounds)
pl.title('LMM')
plt = pl.subplot(2,2,4)
pl.title('LMM')
qqplot(pvalues_lmm)
```

```python
#load first flowering phenotype
#phenotype_query = "(phenotype_ID=='%s')" %  phenotype_
phenotype_names = ['5_FT10']
phenotype_ID =  '5_FT10'
phenotype_query = "(phenotype_ID in %s)" %  str(phenotype_names)

data_flc = dataset.subsample_phenotypes(phenotype_query=phenotype_query,
                                           intersection=True)

sample_relatedness_unnormalized = data_flc.getCovariance(normalize=True,center=True)
sample_relatedness  = sample_relatedness_unnormalized/sample_relatedness_unnormalized.diagonal().mean()

position  = data_flc.getPos()
chrom = 5
pos_start = 13000000
pos_end   = 17000000
idx       = sp.array((position.chrom==chrom) & (pos_start<position.pos) &(position.pos<pos_end),dtype=bool)
idx_start = sp.nonzero(idx)[0].min()
idx_end   = sp.nonzero(idx)[0].max()
snps      = data_flc.getGenotypes()
snps      = snps[:,idx]

snps-= snps.mean(axis=0)
snps/= snps.std(axis=0) 
position = sp.array(position['pos'][idx])

phenotypes,sample_idx = data_flc.getPhenotypes(phenotype_query=phenotype_query,intersection=True) 
phenotype_std = preprocess.rankStandardizeNormal(phenotypes.values)

N = snps.shape[0]
S = snps.shape[1]
P = phenotypes.shape[1]
print "loaded %d samples, %d phenotypes, %s snps" % (N,P,S)
```

```python
# running cross-validation
alphas = 2.**(sp.linspace(-2,-5,10))
alphas = alphas[::-1]
lasso = lmmlasso.LmmLasso(warm_start=True,fit_intercept=False)
MSE_train,MSE_test,W_nonzero = lmmlasso.runCrossValidation(lasso,snps,phenotype_std,alphas,n_folds=10,K=sample_relatedness,verbose=True)
```

```python
# plotting cross-validation results
idx = sp.argmin(MSE_test.mean(axis=0))

pl.figure(figsize=[20,4])
plt = pl.subplot(1,3,1)
plt.plot(sp.log2(alphas),MSE_train.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('training error')
pl.grid(True)

plt = pl.subplot(1,3,2)
plt.plot(sp.log2(alphas),MSE_test.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('test error')
pl.grid(True)

plt = pl.subplot(1,3,3)
plt.plot(sp.log2(alphas),W_nonzero.mean(axis=0),linewidth=2)
pl.axvline(sp.log2(alphas[idx]),color='r')
pl.xlabel('log alpha')
pl.ylabel('number of nonzero coefficients')
pl.grid(True)
```
